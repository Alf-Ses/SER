#### SAVEE

```python
savee_df = pd.read_csv("SAVEE/savee_complete_files.csv")
savee_df
```

```python
set(savee_df.emotion.unique()) - set(ravdess_df.emotion.unique()) - set(emoita_df.emotion.unique())
```

```python
SAVEE_MAPPING = {
    "anger":    "anger",
    "disgust":  "disgust",
    "fear":     "fear",
    "happy":    "happiness",
    "neutral":  "neutral",
    "sadness":  "sadness",
    "surprise": "surprise",
}

savee_df.rename(columns={"emotion": "emotion_def"}, inplace=True)
savee_df["emotion"] = savee_df["emotion_def"].map(SAVEE_MAPPING)
savee_df["split"] = None
savee_df
```

```python
if set(savee_df.emotion.unique()) - set(ravdess_df.emotion.unique()) - set(emoita_df.emotion.unique()) == set():
    print("All emotions mapped correctly.")
else:
    print(set(savee_df.emotion.unique()) - set(ravdess_df.emotion.unique()) - set(emoita_df.emotion.unique()))
```

```python
savee_df.rename(columns={"speaker": "speaker_def"}, inplace=True)
savee_df
```

```python
savee_df["speaker"] = savee_df["speaker_def"].apply(lambda x: f"savee_{x}")
savee_df["split"] = None
savee_df
```

```python
savee_df = savee_df[["file", "speaker", "emotion", "source", "split"]]
savee_df
```

```python
savee_df.to_csv("SAVEE/savee_complete_files.csv", index=False)
```

#### TESS

```python
tess_df = pd.read_csv("TESS/tess_complete_files.csv")
tess_df
```

```python
tess_df.rename(columns={"speaker": "speaker_def"}, inplace=True)
tess_df["speaker"] = tess_df["speaker_def"].apply(lambda x: f"tess_{x}")
tess_df["source"] = "tess"
tess_df["split"] = None
tess_df
```

```python
tess_df = tess_df[["file", "speaker", "emotion", "source", "split"]]
tess_df
```

```python
tess_df.to_csv("TESS/tess_complete_files.csv", index=False)
```

#### Creating train-dev split for training

```python
tmp_dataset = master_df.loc[master_df.split != "test"]
tmp_dataset.split.value_counts()
```

```python
target_train_count = int(len(tmp_dataset) * 0.8)
target_dev_count = len(tmp_dataset) - target_train_count
print(f"Target train count: {target_train_count}.\n Target dev count: {target_dev_count}.")
```

```python
tmp_dataset[tmp_dataset.split == "tmp"].speaker.value_counts()
```

```python
dev_needs = target_dev_count - len(tmp_dataset[tmp_dataset.split == "dev"])
dev_needs
```

```python
candidates = list(tmp_dataset[(tmp_dataset.split == "tmp") & (tmp_dataset.source == "savee")].speaker.unique())
candidates
```

```python
random.seed(4226)

chosen_one = random.choice(candidates)
print(f"Chosen speaker for dev set: {chosen_one}")
```

```python
tmp_dataset.loc[(tmp_dataset.speaker == chosen_one) & (tmp_dataset.split == "tmp"), "split"] = "dev"
tmp_dataset.split.value_counts()
```

```python
tmp_dataset.loc[tmp_dataset.split == "tmp", "split"] = "train"
tmp_dataset.split.value_counts()
```
